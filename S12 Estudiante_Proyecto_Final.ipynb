```python

def revisar_negativos(datos):
    columnas_numericas = datos.select_dtypes(include=['number']).columns
    for columnas in columnas_numericas:
        negativos = (datos[columnas] <= 0).sum()
        print(f'{columnas}:{negativos}')
    print()

        


print('negativos en orders')
revisar_negativos(orders)
print('negativos catalog')
revisar_negativos(catalog)
print('negativos Marketing')
revisar_negativos(marketing)
```

```python
# PARTE 1: Totales del funnel
# ======================

query_totals = '''
SELECT 
    nombre_evento,
    COUNT(DISTINCT id_usuario) as usuarios_unicos
FROM events
GROUP BY nombre_evento
ORDER BY usuarios_unicos DESC;
'''

totals = pd.read_sql(query_totals, con=engine)
totals
```

**Nota:** Se realiza un conteo unico por usuarios que se encuentren agrupados por la columna ***nombre_evento*** esto para ver el volumen general de cada etapa.

```python
# PARTE 2: Conversiones
# ======================

query_conversion = '''

SELECT 

    -- Tasas de conversión
    ROUND(
        COUNT(DISTINCT CASE WHEN nombre_evento = 'add_to_cart' THEN id_usuario END) * 100.0 / 
        COUNT(DISTINCT CASE WHEN nombre_evento = 'first_visit' THEN id_usuario END), 2
    ) AS conversion_add_to_cart,
    
    ROUND(
        COUNT(DISTINCT CASE WHEN nombre_evento = 'select_item' THEN id_usuario END) * 100.0 / 
        COUNT(DISTINCT CASE WHEN nombre_evento = 'add_to_cart' THEN id_usuario END), 2
    ) AS conversion_select_item,

    ROUND(
        COUNT(DISTINCT CASE WHEN nombre_evento = 'begin_checkout' THEN id_usuario END) * 100.0 / 
        COUNT(DISTINCT CASE WHEN nombre_evento = 'select_item' THEN id_usuario END), 2
    ) AS conversion_begin_check_out,

    ROUND(
        COUNT(DISTINCT CASE WHEN nombre_evento = 'add_payment_info' THEN id_usuario END) * 100.0 / 
        COUNT(DISTINCT CASE WHEN nombre_evento = 'begin_checkout' THEN id_usuario END), 2 
    ) AS conversion_add_payment_info,

    ROUND(
        COUNT(DISTINCT CASE WHEN nombre_evento = 'purchase' THEN id_usuario END) * 100.0 / 
        COUNT(DISTINCT CASE WHEN nombre_evento = 'add_payment_info' THEN id_usuario END), 2
    ) AS conversion_purchase,

    ROUND(
        COUNT(DISTINCT CASE WHEN nombre_evento = 'purchase' THEN id_usuario END) * 100 / 
        COUNT(DISTINCT CASE WHEN nombre_evento = 'first_visit' THEN id_usuario END), 2
    ) AS Conversion_Final
    
FROM events
WHERE nombre_evento IN ('first_visit', 'add_to_cart', 'select_item', 'begin_checkout','add_payment_info', 'purchase');


'''

conversion = pd.read_sql(query_conversion, con=engine)
conversion
```

**NOTA:** Se logra apreción que la mayor perdida de usuarios se encuentran en la etapa de ***add_payment_info*** donde solo el ***87.71%*** de los usuarios que habían llegado a al a etapa de ***begin_check_out*** lograron continuar con camino hasta llegar a ***purchase***, se perdio el ***13.29%*** solo entre esas etapas.

## 🔹 Paso 4: Evaluar si los usuarios regresan (retención por cohortes)

**🎯 Objetivo:** Analizar la retención de usuarios para entender si regresan después de registrarse.

**Tablas**

- `users` 
- `user_activity` 

---
1. Se identifica la cohorte de cada usuario según el **mes de registro**.


2. Se calcula la retención semanal: cuántos usuarios **se mantienen activos** en cada semana desde su registro.
   - `retenido_w1`: usuarios activos en la semana 1  
   - `retenido_w2`: usuarios activos en la semana 2  
   - `retenido_w3`: usuarios activos en la semana 3  


3. Se calcula el porcentaje de retención para cada semana, dividiendo los usuarios retenidos entre los clientes iniciales de la cohorte:  
   - `semana_1`: porcentaje de usuarios retenidos en la semana 1  
   - `semana_2`: porcentaje de usuarios retenidos en la semana 2  
   - `semana_3`: porcentaje de usuarios retenidos en la semana 3  

Se revisa que la columna de fecha esté en formato correcto (`DATE`).  
Se realiza la conversión usando: `CAST(fecha_registro AS DATE)`

```python
# Explorar tabla users
# =========================
query_users = '''
SELECT *
FROM users;
'''
users = pd.read_sql(query_users, con=engine)
users.head(3)
```

```python
# Explorar tabla user_activity
# =========================
query_user_activity = '''
SELECT *
FROM user_activity;
'''
user_activity = pd.read_sql(query_user_activity, con=engine)
user_activity.head(10)
```

```python
# Retención por cohortes
# ======================

query_cohort_retention_final = '''
WITH cohortes AS (
    SELECT 
        id_usuario,
        DATE_TRUNC('month', CAST(fecha_registro AS DATE)) as mes_cohorte,
        fecha_registro
    FROM users
),
usuarios_por_cohorte AS (
    SELECT 
        mes_cohorte,
        COUNT(DISTINCT id_usuario) as usuarios_iniciales
    FROM cohortes
    GROUP BY mes_cohorte
),
actividad_semanal AS (
    SELECT 
        c.id_usuario,
        c.mes_cohorte,
        ua.dias_despues_registro,
        CASE 
            WHEN ua.dias_despues_registro BETWEEN 1 AND 7 THEN 'semana_1'
            WHEN ua.dias_despues_registro BETWEEN 8 AND 14 THEN 'semana_2'
            WHEN ua.dias_despues_registro BETWEEN 15 AND 21 THEN 'semana_3'
            WHEN ua.dias_despues_registro BETWEEN 22 AND 28 THEN 'semana_4'
        END as periodo,
        COALESCE(ua.activo, 0) as activo
    FROM cohortes c
    LEFT JOIN user_activity ua ON c.id_usuario = ua.id_usuario
    WHERE ua.dias_despues_registro <= 28
)
SELECT 
    DATE(upc.mes_cohorte),
    upc.usuarios_iniciales AS Usuarios_iniciales,
    COUNT(DISTINCT CASE WHEN periodo = 'semana_1' AND activo = 1 THEN id_usuario END) as retenido_w1,
    COUNT(DISTINCT CASE WHEN periodo = 'semana_2' AND activo = 1 THEN id_usuario END) as retenido_w2,
    COUNT(DISTINCT CASE WHEN periodo = 'semana_3' AND activo = 1 THEN id_usuario END) as retenido_w3,
    COUNT(DISTINCT CASE WHEN periodo = 'semana_4' AND activo = 1 THEN id_usuario END) as retenido_w4
FROM usuarios_por_cohorte upc 
LEFT JOIN actividad_semanal as_ ON upc.mes_cohorte = as_.mes_cohorte
GROUP BY upc.mes_cohorte, upc.usuarios_iniciales
ORDER BY upc.mes_cohorte;
'''

# Ejecutar la consulta
cohorte_final = pd.read_sql(query_cohort_retention_final, con=engine)
cohorte_final
```

```python
# Retención por cohortes
# ======================

query_cohort_retention_final = '''
WITH cohortes AS (
    SELECT 
        id_usuario,
        DATE_TRUNC('month', CAST(fecha_registro AS DATE)) as mes_cohorte,
        fecha_registro
    FROM users
),
usuarios_por_cohorte AS (
    SELECT 
        mes_cohorte,
        COUNT(DISTINCT id_usuario) as usuarios_iniciales
    FROM cohortes
    GROUP BY mes_cohorte
),
actividad_semanal AS (
    SELECT 
        c.id_usuario,
        c.mes_cohorte,
        ua.dias_despues_registro,
        CASE 
            WHEN ua.dias_despues_registro BETWEEN 1 AND 7 THEN 'semana_1'
            WHEN ua.dias_despues_registro BETWEEN 8 AND 14 THEN 'semana_2'
            WHEN ua.dias_despues_registro BETWEEN 15 AND 21 THEN 'semana_3'
            WHEN ua.dias_despues_registro BETWEEN 22 AND 28 THEN 'semana_4'
        END as periodo,
        COALESCE(ua.activo, 0) as activo
    FROM cohortes c
    LEFT JOIN user_activity ua ON c.id_usuario = ua.id_usuario
    WHERE ua.dias_despues_registro <= 28
)
SELECT 
    DATE(upc.mes_cohorte),
    upc.usuarios_iniciales*100/upc.usuarios_iniciales AS Usuarios_iniciales,
    ((COUNT(DISTINCT CASE WHEN periodo = 'semana_1' AND activo = 1 THEN id_usuario END))*100)/ upc.usuarios_iniciales as retenido_w1,
    ((COUNT(DISTINCT CASE WHEN periodo = 'semana_2' AND activo = 1 THEN id_usuario END))*100)/ upc.usuarios_iniciales as retenido_w2,
    ((COUNT(DISTINCT CASE WHEN periodo = 'semana_3' AND activo = 1 THEN id_usuario END))*100)/ upc.usuarios_iniciales as retenido_w3,
    ((COUNT(DISTINCT CASE WHEN periodo = 'semana_4' AND activo = 1 THEN id_usuario END))*100)/ upc.usuarios_iniciales as retenido_w4
FROM usuarios_por_cohorte upc 
LEFT JOIN actividad_semanal as_ ON upc.mes_cohorte = as_.mes_cohorte
GROUP BY upc.mes_cohorte, upc.usuarios_iniciales
ORDER BY upc.mes_cohorte;
'''

# Ejecutar la consulta
cohorte_final = pd.read_sql(query_cohort_retention_final, con=engine)
cohorte_final
```

**Nota:** Se puede apreciar que hay una retención consistente del aproxicamente 40-43% de usuarios activos cada semana, las cohortes presentan estabilidad ya que los porsentajes que se muestran son muy similares entre los diferentes meses evaluados.

## 🔹 Paso 5: Validar si los cambios generan impacto (test estadístico)

🎯 **Objetivo:** Evaluar si la modificación en la UI del checkout impacta la **tasa de conversión de compra**.

---

1. **Analizar el dataset** `experiment_checkout_ui.csv` para identificar la métrica principal **conversion**.
   - La métrica **conversion** es 1 si el usuario completó la compra, 0 si no.    
2. **Plantear la hipótesis estadística**     
3. **Aplicar el test estadístico adecuado** 
4. **Interpretar el resultado**  

---
Hipótesis estadística
   - **H₀ (Hipótesis nula):** Los cambios en el UI del checkout NO impactan en la tasa de conversión 
   - **H₁ (Hipótesis alternativa):**  Los cambios en el UI del checkout SÍ generan una diferencia significativa
   
**Test estadístico:** Z-Test  
**Nivel de significancia alpha:**  α = 0.05

**Nota:** Antes de pasar a las pruebas estadisticas primero se opta por realizar una exploración de datos, para poder determinar si convertir datos, limpiar y corregir lo que sea necesario para hacer una prubea confiable.

```python
conversiones = experiment.groupby('variante')['convirtio'].sum()
totales = experiment.groupby('variante')['convirtio'].count()
print(f'total de conversiones por {conversiones}')
print()
print(f'total de usuarios por {totales}')
```

```python
conversion_resumen = experiment.groupby('variante')['convirtio'].mean()
print(f'Tasa de conversion por {conversion_resumen}')
```

**Nota:** Se utiliza el metodo ***.sum()*** para sumar todos los valroes 1 de la columna ***convirtio***  para saber la cantidad numerica de conversión de los usuarios y ***.count()*** para ver el volumen completo de la muestra de usuarios. Con fines de evaluar la tasa de conversiones de ambas columnas se utiliza el metodo .***mean()***. 

```python
from statsmodels.stats.proportion import proportions_ztest
z_stat, p_value = proportions_ztest(conversiones, totales)

print(f"Estadístico z: {z_stat}")
print(f"Valor p: {p_value}")

alpha = 0.05 
if p_value < alpha:
    print("Rechazamos la hipótesis nula: hay evidencia de una diferencia.")
else:
    print("No rechazamos la hipótesis nula: no hay evidencia suficiente de una diferencia.")
```
